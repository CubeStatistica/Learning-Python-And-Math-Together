```python
x_4 = n.arange(-10, 10)

def f(x):
    import numpy as n
    y = 2*n.cos(0.5*x+1) + n.sin(x) + x/100 #+ (x**3)/5
    return(y)

y_4 = f(x_4)
```

```python
def plot_fn(x_vector,y_vector, mathFunction = None):
    
    import numpy as n
    import matplotlib.pyplot as plt

    xmin = n.min(x_vector)
    xmax = n.max(x_vector)

    ymin = n.min(y_vector)
    ymax = n.max(y_vector)


    plt.style.use('seaborn')
    fig, ax = plt.subplots()
    plt.axis([xmin,xmax,ymin,ymax]) # window size
    plt.plot([xmin,xmax],[0,0],'black' , linestyle = 'dotted') # black x axis
    plt.plot([0,0],[ymin,ymax], 'black', linestyle = 'dotted') # black y axis
    plt.xlabel('x')
    plt.ylabel('y')
    plt.plot(x_vector,y_vector, 'black', marker = "o", label = mathFunction) # Add points on the graph of the function to show the points that were explicitly plotted.
    plt.legend(bbox_to_anchor=(0, 1.02, 1, 0.2), loc="lower left",
                mode="expand", borderaxespad=0, ncol=3, fontsize = 15)
    plt.show()
```

One thing we see immediately to improve our plotting function is to slightly more on the top and the bottom. Therefore, we will modify our plot_fn() to improve it. 

```python
def plot_fn(x_vector,y_vector, mathFunction = None):
    
    import numpy as n
    import matplotlib.pyplot as plt

    xmin = n.min(x_vector) 
    xmax = n.max(x_vector)

    ymin = n.min(y_vector)*(1.10) # We will plot 10% below.
    ymax = n.max(y_vector)*(1.10) # We will plot 10% above.


    plt.style.use('seaborn')
    fig, ax = plt.subplots()
    plt.axis([xmin,xmax,ymin,ymax]) # window size
    plt.plot([xmin,xmax],[0,0],'black' , linestyle = 'dotted') # black x axis
    plt.plot([0,0],[ymin,ymax], 'black', linestyle = 'dotted') # black y axis
    plt.xlabel('x')
    plt.ylabel('y')
    plt.plot(x_vector,y_vector, 'black', marker = "o", label = mathFunction) # Add points on the graph of the function to show the points that were explicitly plotted.
    plt.legend(bbox_to_anchor=(0, 1.02, 1, 0.2), loc="lower left",
                mode="expand", borderaxespad=0, ncol=3, fontsize = 15)
    plt.show()
```

We will now plot this and see how this looks

```python
plot_fn(x, x1)
```

Very nice! This is much better. Now let's include the function as a label in the graph.
