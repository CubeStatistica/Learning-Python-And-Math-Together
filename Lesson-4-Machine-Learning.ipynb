First we import all the python libraries.

```python
import numpy as n
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt


print("All modules loaded")
print(f"Ran code chunk at {dt.datetime.now()}")
```

Recall the main function we have been using:

$f(x) = 2n.cos(0.5x+1) + n.sin(x) + x/100$

And we simply called $y = f(x)$

```python
x = n.arange(-20, 10)

print("x", x)

y = 2*n.cos(0.5*x+1) + n.sin(x) + x/100

print("y", y)
```

Now, remember this function can be separated into three separate components: $x_1, x_2, x_3$. 

Where: 

1. $x_1 = 2*cos(0.5*x+1)$
2. $x_2 = sin(x)$
3. $x_3 = x/100$

You will also remember the code below. Notice that the heart of everything is creating the list, i.e. the array x. This numpy array 'x' is created using n.arange() function. Also, this list object in math is called a vector.

```python
x1 = 2*n.cos(0.5*x+1)
x2 = n.sin(x)
x3 = x/100

print('x1', x1, '\n\n', 'x2', x2, '\n\n', 'x3', x3)
```

The above is not a great way to view the three vectors.

We will put this into a dataframe and print it.

Easiest way to make a dataframe using lists of numbers in python, also known as vectors or arrays is to first create a python dictionary. 

```python
X = dict(x1 = x1, x2 = x2, x3 = x3)
print(X)
```

```python
df = pd.DataFrame(X)
df
```

This is how you will typically receive data. 

These $x_1, x_2, x_3$ can be thought of as variables, as they are varying. So we can rename these to variable1, variable2, variable3.

Further, when you get a dataset you typically also get a variable of interest. It is the main thing you are interested in, for example, the price of a stock, or the height of a person, etc. It is also called outcome variable. And we can think of y as our outcome variable and you would receive this in the dataset. 

The way you can add columns to your dataframe is as follows:

```python
df['y'] = y

df
```

Now typically, when you receive such a data. Your goal would be to find how the variables are related to the outcome variable.

In our case, if we simply do $x_1 + x_2 + x_3$ it will be equal to y. It is possible we also had the variable $x$ also given to us. So will will add that to the dataset.



```python
df['x'] = x # Here we have added the x variable. However, it was add at the end.
# Want to added it at the beginning. We show how to do taht below.
print(df)
first_col = df.pop('x')
df.insert(0, 'x', first_col)
df
```
